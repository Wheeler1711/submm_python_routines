```python
import os

import numpy as np
import matplotlib.pyplot as plt

from submm.KIDs.res import fitting as res_fit
from submm.sample_data.abs_paths import abs_path_sample_data
from importlib import reload
from submm.KIDs import noise_theory as nt
```

# Single Resonator Fitting
## Load the Sample Data

```python
data_path = os.path.join(abs_path_sample_data,"temps_tls.npy")
temps = np.load(data_path)
data_path = os.path.join(abs_path_sample_data,"f_tls.npy")
f = np.load(data_path)
f_Hz = f*10**9
data_path = os.path.join(abs_path_sample_data,"z_tls.npy")
z = np.load(data_path)
```

### Plot the first resonator data

```python
for i in range(0,z.shape[2]):
    plt.plot(f[:,0,i],20*np.log10(np.abs(z[:,0,i])),label = str(temps[i])+"mK")
    
plt.ylabel("Power (dB)")
plt.xlabel("Frequency (GHz)")
plt.legend()
plt.show()
```

### Fit the data to extract resonator frequency

```python
f_centers = []
Qis = []
for i in range(0,z.shape[2]):
    fit = res_fit.fit_nonlinear_iq(f_Hz[:,0,i],z[:,0,i])
    f_centers.append(fit.result.fr)
    Qis.append(fit.result.Qi)
    fit.plot()
    
f_centers = np.asarray(f_centers)
Qis = np.asarray(Qis)
```

### Plot the fit results

```python
# convert to df/f units
df_over_f = (f_centers-f_centers[0])/f_centers[0]
plt.figure()
plt.plot(temps,f_centers/10**9)
plt.ylabel("Frequency (GHz)")
plt.xlabel("Temperature (mK)")
plt.figure()
plt.plot(temps,df_over_f)
plt.ylabel("Frequency (GHz)")
plt.xlabel("Temperature (mK)")
plt.figure()
plt.plot(temps,Qis)
plt.ylabel("Qi")
plt.xlabel("Temperature (mK)")
```
